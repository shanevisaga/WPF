```python
import numpy as np
import pandas as pd
import xarray as xr

import matplotlib as mp
import matplotlib.pyplot as plt
import matplotlib.colors as colors


from datetime import datetime,timezone 
```

```python
ds = xr.open_dataset('threshold_0.2/second_harmonic_fft_withTC.nc')
```

```python
pa = ds.PA
```

```python
df = pa.to_dataframe().reset_index().dropna()
df = df.groupby(['lat','lon','time']).agg({'PA': ['mean']}).droplevel(axis=1, level=0).reset_index()
```

```python
df
```

```python
df.pivot(index=['lon','lat'], columns='time', values='mean').reset_index().to_csv('pa_withTC.csv')
```

```python
df.pivot(index=['lon','lat'], columns='time', values='mean').reset_index()
```

```python
pf = ds.PF
df = pf.to_dataframe().reset_index().dropna()
df = df.groupby(['lat','lon','time']).agg({'PF': ['mean']}).droplevel(axis=1, level=0).reset_index()
df.pivot(index=['lon','lat'], columns='time', values='mean').reset_index().to_csv('pf_withTC.csv')

pi = ds.PI
df = pi.to_dataframe().reset_index().dropna()
df = df.groupby(['lat','lon','time']).agg({'PI': ['mean']}).droplevel(axis=1, level=0).reset_index()
df.pivot(index=['lon','lat'], columns='time', values='mean').reset_index().to_csv('pi_withTC.csv')

```
